# Data Quality Summary

## Overview

An initial data quality assessment was performed on the mutual fund datasets provided for the project.

## Observations

* All datasets were successfully loaded into Pandas DataFrames without any errors.
* Dataset dimensions and data types were inspected using `.shape` and `.dtypes`.
* Unique fund houses, categories, sub-categories, and risk categories were successfully identified.
* The `amfi_code` column was found to be present and can be used as a unique identifier for mutual fund schemes.
* AMFI codes from `fund_master` were validated against `nav_history` to ensure consistency across datasets.
* No duplicate records were found in the datasets examined during the initial exploration.
* The `04_monthly_sip_inflows.csv` dataset contains 12 missing values in the `yoy_growth_pct` column.
* These missing values are likely due to insufficient previous-year data required for calculating year-over-year growth percentages.
* Apart from the above missing values, no major data quality issues or inconsistencies were observed.
* Overall, the datasets appear suitable for further analysis and dashboard development.

## Conclusion

The overall quality of the data is good. Minor missing values were identified in the `yoy_growth_pct` column; however, the datasets are largely consistent and ready for subsequent stages such as data cleaning, transformation, exploratory data analysis, and visualization.
